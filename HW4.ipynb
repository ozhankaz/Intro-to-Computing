### Question 4: Plot a figure with multiple lines, as shown in figure 2, by simply adding more plt.plot( ) functions.

```python
y1 = np.array([2, 4, 6, 7])
y2 = np.array([7, 9, 11, 12])
y3 = np.array([1, 2, 3, 5])

plt.plot(y1)
plt.plot(y2)
plt.plot(y3)

plt.xlabel("X-Values")
plt.ylabel("Y-Values")

plt.show()
```

### Question 5: Practice subplots( ) function, design your own subplot. Using whatever we learned so far, you can draw multiple plots in one figure. Do not copy our class practice code.

```python
x = np.arange(7, 50)
y1 = 1/np.sqrt(x)
y2 = np.tan(x)
y3 = x ** 2
```

```python
fig = plt.figure(figsize=(10, 10)) #increases subplot size

plt1 = fig.add_subplot(212)
plt.plot(x, y1)

plt.xticks(fontsize = 14) # increase tick font size for y1 plot 
plt.yticks(fontsize = 14)

plt2 = fig.add_subplot(221)
plt.plot(x, y2)

plt.xticks(fontsize = 14) # increase tick font size for y2 plot 
plt.yticks(fontsize = 14)

plt3 = fig.add_subplot(222)
plt.plot(x, y3)

plt.xticks(fontsize = 14) # increase tick font size for y3 plot 
plt.yticks(fontsize = 14)

plt.show()
```
